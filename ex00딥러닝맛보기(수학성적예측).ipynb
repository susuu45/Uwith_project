### 신경망 모델 만들기
1. 신경망 구조 설계
2. 신경망 학습과 평가 방법 설정
3. 학습 및 학습 과정 시각화
4. 모델 평가 및 예측

```python
# 단일 특성 분석 모델 만들기
# 1개의 특성을 이용해서 정답을 예측해보자
```

```python
# 신경망 구현 도구 불러오기
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Activation
# Sequential() : 신경망 모델의 뼈대를 구축해주는 함수 / 각자의 층을 선형으로 연결해주는 역할을 담당
# 층을 위에서 아래로 쌓는 구조 - 각 층을 쌇아서 모델을 만들어 주는 기능
# Dense() : 신경망의 층(뉴런의 묶음)을 설정하고 구성하는 함수
# 층마다 모든 방이 서로 연결 - 입력된 정보가 고르게 전달될 수 있도록 뉴럭을 설정하는 기능
# Activation() : 정보를 판단하고 판단한 결과에 따라 값을 변경해주는 함수(활성화 함수)
# -> 신호를 보낼지 말지 결정하는 스위치 / 필터와 같은 역할
# 출입 통제 시스템 - 보안 시스템이 판단했을때 이 방은 필요가 없어! -> 정보 차단
```

```python
# 모델 구현
# 모델 뼈대 설정
model = Sequential()

# 층 추가
# add() : 층을 쌓아주는 (추가하는) 함수
# Dense : 층의 구조를 결정하는 함수
# units : 뉴런의 갯수를 설정하는 매개변수
# input_dim : 입력할 데이터의 특징 갯수

# 입력층 구성
model.add(Dense(units = 4, input_dim = 1))
# Sequential이라는 뼈대에 뉴런의 갯수 4개 / 특성을 1개 입력받는 층을 추가 시켜주세요
model.add(Activation('sigmoid'))
# Sequentioal이라는 뼈대에 Activaition(활성화 함수)로 'sgmoid'함수를 추가해주세요

# 출력층 구성
model.add(Dense(units = 1))
# Sequential이라는 뼈대에 뉴런의 갯수가 1개인 층을 추가시켜주세요
# Activation 생략
```

```python
model.summary()
```

```python
# 2. compile - 학습 / 평가 방법 설정
# compile() : 앞서 만든 모델이 효과적으로 구현될 수 있도록 환경설정을 해주면서 컴파일을 진행하는 함수
# 모델을 학습시키기 전에 손실함수와 최적화 방법(경사 하강법과 같은 친구 설정), 성능 평가지 지표를 설정하는 함수
model.compile(loss = 'mse', # 손실 함수 : 오차를 측정하는 도구 / 현재는 회귀분석을 진행하기 때문에 평균 제곱 오차 설정
              optimizer = 'SGD', # 최적화 함수 : 모델의 성능이 잘 나오도록 최적화 시키는 방법을 설정 -> 고급 경사하강법
              # metrics : 평가 지표 설명하는 부분 / 모델이 학습이 잘 되었는지 성능 평가시에 사용
              )
```

```python
# 2. 모델 학습
# fit()
model.fit(X_train['studytime'], y_train, epochs=300)
# epochs : 학습 반복 횟수
```

```python
# 예측
model.predict(X_test['studytime'])
```

```python
# 모델 성능 평가
model.evaluate(X_test['studytime'], y_test)
```

```python
# 특성 두개를 입력받아 분석하는 모델 만들기
# 모델 구현
# 모델 뼈대 설정
model1 = Sequential()

# 층 추가
# add() : 층을 쌓아주는 (추가하는) 함수
# Dense : 층의 구조를 결정하는 함수
# units : 뉴런의 갯수를 설정하는 매개변수
# input_dim : 입력할 데이터의 특징 갯수

# 입력층 구성
model1.add(Dense(units = 4, input_dim = 2))
# Sequential이라는 뼈대에 뉴런의 갯수 4개 / 특성을 2개 입력받는 층을 추가 시켜주세요
model1.add(Activation('sigmoid'))
# Sequentioal이라는 뼈대에 Activaition(활성화 함수)로 'sgmoid'함수를 추가해주세요

# 출력층 구성
model1.add(Dense(units = 1))
# Sequential이라는 뼈대에 뉴런의 갯수가 1개인 층을 추가시켜주세요
# Activation 생략
```

```python
# 2. compile - 학습 / 평가 방법 설정
# compile() : 앞서 만든 모델이 효과적으로 구현될 수 있도록 환경설정을 해주면서 컴파일을 진행하는 함수
# 모델을 학습시키기 전에 손실함수와 최적화 방법(경사 하강법과 같은 친구 설정), 성능 평가지 지표를 설정하는 함수
model1.compile(loss = 'mse', # 손실 함수 : 오차를 측정하는 도구 / 현재는 회귀분석을 진행하기 때문에 평균 제곱 오차 설정
              optimizer = 'SGD', # 최적화 함수 : 모델의 성능이 잘 나오도록 최적화 시키는 방법을 설정 -> 고급 경사하강법
              # metrics : 평가 지표 설명하는 부분 / 모델이 학습이 잘 되었는지 성능 평가시에 사용
              )
```

```python
# 2. 모델 학습
# fit()
model1.fit(X_train[['studytime','traveltime']], y_train, epochs=300)
# epochs : 학습 반복 횟수
```
